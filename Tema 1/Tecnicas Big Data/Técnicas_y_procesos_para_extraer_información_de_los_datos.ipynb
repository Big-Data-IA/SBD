# Análisis Estadístico

```python
# Importar bibliotecas
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Crear dos conjuntos de datos ficticios
np.random.seed(0)
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 1, 100)

# Resumen estadístico
data1_mean = np.mean(data1)  # Cálculo de la media
data2_mean = np.mean(data2)

data1_median = np.median(data1)  # Cálculo de la mediana
data2_median = np.median(data2)

data1_std = np.std(data1)  # Cálculo de la desviación estándar
data2_std = np.std(data2)

data1_skewness = stats.skew(data1)  # Cálculo de la asimetría
data2_skewness = stats.skew(data2)

data1_kurtosis = stats.kurtosis(data1)  # Cálculo de la curtosis
data2_kurtosis = stats.kurtosis(data2)

# Gráficos de los datos
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data1, kde=True, color='blue', label='Data 1')
sns.histplot(data2, kde=True, color='red', label='Data 2')
plt.title('Distribución de los Datos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()

plt.subplot(1, 2, 2)
sns.boxplot(data=[data1, data2], palette=['blue', 'red'])
plt.title('Diagrama de Caja y Bigotes')
plt.xticks([0, 1], ['Data 1', 'Data 2'])

plt.show()

# Imprimir estadísticas
print("Estadísticas para el primer conjunto de datos:")
print(f"Media: {data1_mean:.4f}")
print(f"Mediana: {data1_median:.4f}")
print(f"Desviación Estándar: {data1_std:.4f}")
print(f"Asimetría (Skewness): {data1_skewness:.4f}")
print(f"Curtosis (Kurtosis): {data1_kurtosis:.4f}")

print("\nEstadísticas para el segundo conjunto de datos:")
print(f"Media: {data2_mean:.4f}")
print(f"Mediana: {data2_median:.4f}")
print(f"Desviación Estándar: {data2_std:.4f}")
print(f"Asimetría (Skewness): {data2_skewness:.4f}")
print(f"Curtosis (Kurtosis): {data2_kurtosis:.4f}")

# Prueba de normalidad de Anderson-Darling
result1 = stats.anderson(data1)  # Prueba de normalidad para Data 1
result2 = stats.anderson(data2)  # Prueba de normalidad para Data 2

print("\nPrueba de Normalidad (Anderson-Darling):")
print(f"Estadístico para Data 1: {result1.statistic:.4f}")
print(f"Valores críticos para Data 1: {result1.critical_values}")
print(f"Estadístico para Data 2: {result2.statistic:.4f}")
print(f"Valores críticos para Data 2: {result2.critical_values}")

# Prueba de hipótesis (prueba t de Student)
t_stat, p_value = stats.ttest_ind(data1, data2)  # Prueba t de Student entre los dos conjuntos de datos

print("\nPrueba de hipótesis (t de Student):")
print(f"Estadístico t: {t_stat:.4f}")
print(f"Valor p: {p_value:.4f}")

# Interpretación del resultado de la prueba de hipótesis
alpha = 0.05  # Nivel de significancia
if p_value < alpha:
    print("Hay una diferencia significativa entre los dos conjuntos de datos.")
else:
    print("No hay una diferencia significativa entre los dos conjuntos de datos.")
```

# Aprendizaje Automático (Machine Learning)

```python
# Importar bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, mean_squared_error
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neural_network import MLPClassifier

# Cargar el conjunto de datos de Iris
iris = datasets.load_iris()
X, y = iris.data, iris.target

# Partir el conjunto de datos en datos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Clasificación utilizando SVM
clf = SVC()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# Mostrar resultados de la clasificación
print("Resultados de la clasificación:")
print(classification_report(y_test, y_pred))

# Regresión para predecir la longitud del pétalo
petal_length = iris.data[:, 2]
sepal_length = iris.data[:, 0]

# Entrenar un modelo de regresión
regression_model = np.polyfit(sepal_length, petal_length, 1)
petal_length_pred = np.polyval(regression_model, sepal_length)

# Calcular el error de regresión (RMSE)
rmse = np.sqrt(mean_squared_error(petal_length, petal_length_pred))

# Mostrar el resultado de la regresión
print(f"Error de regresión (RMSE): {rmse:.4f}")

# Agrupación utilizando K-Means
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
cluster_labels = kmeans.labels_

# Mostrar etiquetas de clústeres K-Means
print("Etiquetas de clústeres K-Means:")
print(cluster_labels)

# Procesamiento de lenguaje natural (NLP)
text_data = ["Este es un buen producto.", "No me gustó el servicio.", "El envío fue rápido y eficiente."]
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(text_data)

# Mostrar la matriz TF-IDF
print("Matriz TF-IDF:")
print(tfidf_matrix.toarray())

# Redes neuronales para clasificación de opiniones
reviews = ["Buena película.", "Mala película.", "Excelente actuación.", "Decepcionante", "Brutal", "No me gustó", "Increible"]
labels = [1, 0, 1, 0, 1, 0, 1]

vectorizer = TfidfVectorizer()
X_reviews = vectorizer.fit_transform(reviews)

nn_classifier = MLPClassifier(hidden_layer_sizes=(5,), max_iter=5000)
nn_classifier.fit(X_reviews, labels)

# Predicción de opiniones
new_reviews = ["La película fue asombrosa.", "Muy decepcionante."]
X_new_reviews = vectorizer.transform(new_reviews)
predicted_labels = nn_classifier.predict(X_new_reviews)

# Mostrar las predicciones de opiniones
print("Predicciones de opiniones:")
for review, label in zip(new_reviews, predicted_labels):
    sentiment = "positiva" if label == 1 else "negativa"
    print(f"Opinión: '{review}' - Sentimiento: {sentiment}")
```

# Minería de datos

```python
# Importar bibliotecas
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns

# URL del conjunto de datos en GitHub
url = "https://raw.githubusercontent.com/nsethi31/Kaggle-Data-Credit-Card-Fraud-Detection/master/creditcard.csv"

# Cargar el conjunto de datos desde GitHub
data = pd.read_csv(url,nrows=40)

# Explorar los primeros registros del conjunto de datos
print("Primeros registros del conjunto de datos:")
print(data.head())

# Reglas de asociación: Identificar patrones y relaciones entre elementos
basket = data.iloc[:, 1:-1]  # Seleccionar columnas relevantes para las reglas de asociación
basket_encoded = basket.applymap(lambda x: 1 if x > 0 else 0)
frequent_itemsets = apriori(basket_encoded, min_support=0.9, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Mostrar reglas de asociación
print("\nReglas de asociación:")
print(rules.head(10))

# Segmentación: Dividir datos en grupos homogéneos
X = data.drop(["Time", "Amount", "Class"], axis=1)  # Seleccionar columnas relevantes para la segmentación
kmeans = KMeans(n_clusters=2, random_state=0)
data["Cluster"] = kmeans.fit_predict(X)

# Visualización de la segmentación
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x="V14", y="V17", hue="Cluster", palette="Dark2")
plt.title("Segmentación de datos de tarjetas de crédito")
plt.show()

# Detección de anomalías: Identificar valores atípicos
X_anomaly = data.drop(["Time", "Amount", "Class"], axis=1)  # Seleccionar columnas relevantes para la detección de anomalías
clf = IsolationForest(contamination=0.02, random_state=0)  # Contaminación estimada
data["Anomaly"] = clf.fit_predict(X_anomaly)

# Visualización de anomalías
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x="V14", y="V17", hue="Anomaly", palette="Set1")
plt.title("Detección de anomalías en datos de tarjetas de crédito")
plt.show()

# Conteo de anomalías
anomalies = data[data["Anomaly"] == -1]
num_anomalies = len(anomalies)

print(f"Número de anomalías detectadas: {num_anomalies}")
```

# Minería de texto

```python
# Importar bibliotecas
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
from sklearn.decomposition import LatentDirichletAllocation
import requests
from bs4 import BeautifulSoup

# Cargar el conjunto de datos "20 Newsgroups"
newsgroups = fetch_20newsgroups(subset='train', shuffle=True, random_state=42, remove=('headers', 'footers', 'quotes'))

# Explorar los primeros documentos del conjunto de datos
print("Primeros documentos del conjunto de datos:")
for i in range(3):
    print(f"Documento #{i + 1}:\n{newsgroups.data[i*2]}\n")

# Clasificación de documentos: Categorizar documentos en diferentes temas
X = newsgroups.data
y = newsgroups.target

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear un modelo de clasificación (usamos un clasificador Naive Bayes en este ejemplo)
vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

classifier = MultinomialNB()
classifier.fit(X_train_tfidf, y_train)
y_pred = classifier.predict(X_test_tfidf)

# Mostrar resultados de la clasificación
print("\nResultados de la clasificación:")
print(classification_report(y_test, y_pred))
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión: {accuracy:.4f}")

# Análisis de temas: Identificar los temas principales en el conjunto de documentos
num_topics = 5
tf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
tf = tf_vectorizer.fit_transform(newsgroups.data)
lda = LatentDirichletAllocation(n_components=num_topics, max_iter=5, learning_method='online', learning_offset=50., random_state=0)
lda.fit(tf)

# Imprimir los temas principales
print("\nTemas principales identificados:")
for topic_idx, topic in enumerate(lda.components_):
    print(f"Tema #{topic_idx + 1}:")
    print(" ".join([tf_vectorizer.get_feature_names_out()[i] for i in topic.argsort()[:-10 - 1:-1]]))

# Búsqueda de información: Extraer información relevante en documentos
def search_documents(query, num_results=5):
    # Simulamos una búsqueda en el conjunto de documentos
    # En un escenario real, usarías motores de búsqueda más avanzados
    search_results = []
    for i, document in enumerate(newsgroups.data):
        if query.lower() in document.lower():
            search_results.append(f"Documento #{i + 1}:\n{document}\n")
            if len(search_results) >= num_results:
                break

    print(f"\nResultados de búsqueda para '{query}':")
    if search_results:
        for result in search_results:
            print(result)
    else:
        print("No se encontraron documentos que coincidan con la búsqueda.")

# Búsqueda de información en documentos
search_query = "Python programming"
search_documents(search_query)
```

# Procesamiento de imágenes

```python
# Importar bibliotecas
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Cargar un clasificador pre-entrenado para la detección de rostros
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Cargar una imagen
image_path = "/content/sample_data/lena.jpg"
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


# Detección de rostros en la imagen
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))

# Dibujar rectángulos alrededor de los rostros detectados
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Mostrar la imagen con los rostros detectados
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Realizar reconocimiento de patrones (ejemplo de detección de bordes)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, threshold1=100, threshold2=200)

# Mostrar la imagen con los bordes detectados
cv2_imshow(edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Instalar BeautifulSoup y Scrapy si no están instalados
!pip install beautifulsoup4 scrapy

# Importar las bibliotecas necesarias
import requests
from bs4 import BeautifulSoup
import scrapy
from scrapy.crawler import CrawlerProcess

# Definir una URL de muestra (sitio web ficticio de noticias)
url = "https://elpais.com/"

# Utilizar BeautifulSoup para extraer datos de la página
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Imprimir el título de la página
title = soup.title.string
print(f"Título de la página: {title}\n")

# Extraer y mostrar los títulos de las noticias en la página
news_headlines = soup.find_all("h2")
print("Títulos de noticias:")
for headline in news_headlines:
    print(headline.text)

# Crear una araña (spider) de Scrapy para extraer datos
class NewsSpider(scrapy.Spider):
    name = "news_spider"
    start_urls = [url]

    def parse(self, response):
        # Extraer y mostrar los títulos de las noticias en la página
        news_headlines = response.css("h2::text").getall()
        print("\nTítulos de noticias (Scrapy):")
        for headline in news_headlines:
            print(headline)

# Ejecutar la araña de Scrapy
process = CrawlerProcess()
process.crawl(NewsSpider)
process.start()
```
